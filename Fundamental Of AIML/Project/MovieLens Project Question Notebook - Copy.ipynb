### 4. Find the number of movies per genre using the item data - 2.5 marks

```python
#your code here
genres= ["unknown","Action","Adventure","Animation","Childrens","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western"]

genrescount = []
for col in movies.columns:
    
    if col in genres:
        tmp_series = movies[col].value_counts()
        genrescount.append(tmp_series)
df_value_counts = pd.concat(genrescount, axis=1)
df_value_counts
```
